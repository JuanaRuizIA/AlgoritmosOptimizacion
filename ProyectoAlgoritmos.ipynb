### 6. Diseña un algoritmo para resolver el problema por fuerza bruta

DISEÑO CONCEPTUALPrincipio: Generar todas las combinaciones posibles de asignación actor-toma y seleccionar la que minimice el número de actores únicos.

```
def brute_force_doblaje(availability_matrix):
    """
    Resuelve el problema de doblaje por fuerza bruta

    Args:
        availability_matrix: Matriz 30x10 (tomas x actores)
                           1 = disponible, 0 = no disponible

    Returns:
        tuple: (mejor_solucion, actores_minimos, total_evaluaciones)
    """

    # Variables de control
    best_solution = None
    min_actors = float('inf')
    total_evaluations = 0

    # Actores disponibles por toma
    available_by_take = []
    for take in range(len(availability_matrix)):
        available_actors = []
        for actor in range(10):
            if availability_matrix[take][actor] == 1:
                available_actors.append(actor)
        available_by_take.append(available_actors)

    print("ALGORITMO FUERZA BRUTA")
    print("=========================================")
    print(f"Actores disponibles por toma: {[len(actors) for actors in available_by_take]}")

    # Total de combinaciones
    total_combinations = 1
    for actors in available_by_take:
        total_combinations *= len(actors)
    print(f"Total combinaciones a evaluar: {total_combinations:,}")

    # Todas las combinaciones posibles
    def generate_all_solutions(take_index, current_solution):
        nonlocal best_solution, min_actors, total_evaluations

        # Caso base: solucion completa
        if take_index == len(availability_matrix):
            total_evaluations += 1

            # Funcion objetivo
            unique_actors = len(set(current_solution))

            # Mejor solucion
            if unique_actors < min_actors:
                min_actors = unique_actors
                best_solution = current_solution.copy()
                print(f"Nueva mejor solucion encontrada: {min_actors} actores (eval #{total_evaluations:,})")

            # Mostrar progreso cada millon de evaluaciones
            if total_evaluations % 1000000 == 0:
                print(f"Progreso: {total_evaluations:,} evaluaciones completadas")

            return

        # Caso recursivo: probar todos los actores disponibles para esta toma
        for actor in available_by_take[take_index]:
            current_solution.append(actor)
            generate_all_solutions(take_index + 1, current_solution)
            current_solution.pop()

    # Busqueda exhaustiva
    import time
    start_time = time.time()

    generate_all_solutions(0, [])

    end_time = time.time()
    execution_time = end_time - start_time

    # Resultados
    print(f"\nRESULTADOS FINALES:")
    print(f"Mejor solucion: {best_solution}")
    print(f"Actores minimos: {min_actors}")
    print(f"Total evaluaciones: {total_evaluations:,}")
    print(f"Tiempo ejecucion: {execution_time:.2f} segundos")
    print(f"Evaluaciones/segundo: {total_evaluations/execution_time:,.0f}")

    return best_solution, min_actors, total_evaluations


# Funcion auxiliar para validar solucion
def validate_solution(solution, availability_matrix):
    """Verifica que la solucion respete las restricciones"""
    for take, actor in enumerate(solution):
        if availability_matrix[take][actor] == 0:
            return False
    return True


# Funcion objetivo
def objective_function(solution):
    """Calcula numero de actores unicos"""
    return len(set(solution))
```

### 7. (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

FUERZA BRUTA CON PODA INTELIGENTE. <br>
El algoritmo de fuerza bruta puro evalúa 1.65 × 10¹⁴ combinaciones sin ninguna optimización, incluyendo:

- Soluciones que ya superan el mínimo encontrado
- Ramas que no pueden mejorar el resultado
- Evaluaciones redundantes de soluciones parciales inviables

```
def brute_force_improved(availability_matrix):
    """
    Algoritmo de fuerza bruta mejorado con poda inteligente
    """

    best_solution = None
    min_actors = float('inf')
    total_evaluations = 0
    pruned_branches = 0

    num_takes = len(availability_matrix)

    # MEJORA 1: Preprocesa y ordena tomas por restriccion
    available_by_take = []
    take_order = []

    for take in range(num_takes):
        available_actors = []
        for actor in range(10):
            if availability_matrix[take][actor] == 1:
                available_actors.append(actor)
        available_by_take.append(available_actors)
        take_order.append((take, len(available_actors)))

    # Ordena tomas: menos actores disponibles primero
    take_order.sort(key=lambda x: x[1])
    ordered_takes = [take for take, _ in take_order]

    print("ALGORITMO FUERZA BRUTA MEJORADO")
    print("==============================")
    print(f"Tomas a procesar: {num_takes}")
    print(f"Orden de tomas optimizado: {ordered_takes}")
    print(f"Restricciones por toma: {[len(available_by_take[t]) for t in ordered_takes]}")

    # Cota inferior teorica
    theoretical_min = estimate_theoretical_minimum(available_by_take, num_takes)
    print(f"Minimo teorico estimado: {theoretical_min} actores")

    def generate_solutions_with_pruning(take_idx, current_solution, current_actors_set):
        nonlocal best_solution, min_actors, total_evaluations, pruned_branches

        # MEJORA 2: Poda por cota superior
        current_actors_count = len(current_actors_set)
        if current_actors_count >= min_actors:
            pruned_branches += 1
            return

        # MEJORA 3: Poda por estimacion optimista
        remaining_takes = len(ordered_takes) - take_idx
        if remaining_takes > 0:
            optimistic_estimate = estimate_minimum_new_actors(
                take_idx, ordered_takes, available_by_take, current_actors_set
            )
            if current_actors_count + optimistic_estimate >= min_actors:
                pruned_branches += 1
                return

        # Caso base: solucion completa
        if take_idx == len(ordered_takes):
            total_evaluations += 1
            unique_actors = len(current_actors_set)

            if unique_actors < min_actors:
                min_actors = unique_actors
                best_solution = current_solution.copy()
                print(f"Nueva mejor solucion: {min_actors} actores (eval #{total_evaluations:,}, podas: {pruned_branches:,})")

                # MEJORA 4: Deteccion de optimo global
                if min_actors == theoretical_min:
                    print("OPTIMO GLOBAL ENCONTRADO - Terminando busqueda")
                    return

            return

        # Caso recursivo con ordenacion dinamica de valores
        current_take = ordered_takes[take_idx]
        available_actors = available_by_take[current_take]

        # MEJORA 5: Ordena actores por preferencia (actores ya usados primero)
        preferred_actors = []
        new_actors = []

        for actor in available_actors:
            if actor in current_actors_set:
                preferred_actors.append(actor)
            else:
                new_actors.append(actor)

        # Probar primero actores existentes, luego nuevos
        ordered_actors = preferred_actors + new_actors

        for actor in ordered_actors:
            current_solution.append(actor)
            new_actors_set = current_actors_set | {actor}

            generate_solutions_with_pruning(take_idx + 1, current_solution, new_actors_set)

            current_solution.pop()

            # MEJORA 6: Poda temprana si ya encontramos optimo
            if min_actors == theoretical_min:
                break

    #Busqueda mejorada
    import time
    start_time = time.time()

    generate_solutions_with_pruning(0, [], set())

    end_time = time.time()
    execution_time = end_time - start_time

    # Estadisticas finales
    total_nodes = total_evaluations + pruned_branches
    pruning_efficiency = (pruned_branches / total_nodes) * 100 if total_nodes > 0 else 0

    print(f"\nRESULTADOS ALGORITMO MEJORADO:")
    print(f"Mejor solucion: {best_solution}")
    print(f"Actores minimos: {min_actors}")
    print(f"Evaluaciones completas: {total_evaluations:,}")
    print(f"Ramas podadas: {pruned_branches:,}")
    print(f"Eficiencia de poda: {pruning_efficiency:.2f}%")
    print(f"Tiempo ejecucion: {execution_time:.2f} segundos")
    print(f"Speedup vs fuerza bruta: {estimate_speedup(pruned_branches):.1f}x")

    return best_solution, min_actors, total_evaluations


def estimate_theoretical_minimum(available_by_take, num_takes):
    """Estima el minimo teorico de actores necesarios"""
    all_actors = set(range(10))
    min_coverage = float('inf')

    for actor in all_actors:
        coverage = sum(1 for take_actors in available_by_take if actor in take_actors)
        if coverage > 0:
            min_coverage = min(min_coverage, num_takes // coverage + (1 if num_takes % coverage else 0))

    return max(1, min_coverage // 3)


def estimate_minimum_new_actors(take_idx, ordered_takes, available_by_take, current_actors):
    """Estima minimo de actores nuevos necesarios para tomas restantes"""
    remaining_takes = ordered_takes[take_idx:]
    uncovered_takes = []

    for take in remaining_takes:
        available = available_by_take[take]
        if not any(actor in current_actors for actor in available):
            uncovered_takes.append(take)

    if not uncovered_takes:
        return 0

    return max(1, len(uncovered_takes) // 5)


def estimate_speedup(pruned_branches):
    """Estima el speedup conseguido por la poda"""
    if pruned_branches == 0:
        return 1.0

    average_subtree_size = 1000000
    evaluations_saved = pruned_branches * average_subtree_size
    original_evaluations = 1.65e14

    return min(original_evaluations / max(1, original_evaluations - evaluations_saved), 10000)


test_matrix = availability_matrix[:8]
resultado = brute_force_improved(test_matrix)

```

### 8. (*)Calcula la complejidad del algoritmo

```
import math
import time
import numpy as np

def calculate_complexity_metrics(availability_matrix):
    """
    Calcula metricas de complejidad

    Args:
        availability_matrix: Matriz de disponibilidad n x 10

    Returns:
        dict: Metricas de complejidad calculadas
    """

    num_takes = len(availability_matrix)

    print("ANALISIS DE COMPLEJIDAD DEL ALGORITMO")
    print("====================================")

    # 1. Complejidad teorica (peor caso)
    actors_per_take = []
    total_combinations = 1

    for take in range(num_takes):
        available_count = sum(availability_matrix[take])
        actors_per_take.append(available_count)
        total_combinations *= available_count

    print(f"Numero de tomas: {num_takes}")
    print(f"Actores disponibles por toma: {actors_per_take}")
    print(f"Promedio actores por toma: {np.mean(actors_per_take):.2f}")
    print(f"Minimo actores por toma: {min(actors_per_take)}")
    print(f"Maximo actores por toma: {max(actors_per_take)}")

    # 2. Complejidad temporal teorica
    print(f"\nCOMPLEJIDID TEMPORAL:")
    print(f"Combinaciones totales (peor caso): {total_combinations:,}")
    print(f"Notacion cientifica: {total_combinations:.2e}")
    print(f"Orden de magnitud: O({np.mean(actors_per_take):.1f}^{num_takes})")

    # 3. Estimacion con factores de poda
    pruning_factors = [0.0, 0.5, 0.8, 0.9, 0.95, 0.99]

    print(f"\nESTIMACIONES CON PODA:")
    print("Tasa Poda | Nodos Visitados | Factor Mejora | Tiempo Estimado")
    print("-" * 65)

    for pruning_rate in pruning_factors:
        effective_combinations = total_combinations * (1 - pruning_rate)
        improvement_factor = total_combinations / max(effective_combinations, 1)
        estimated_hours = effective_combinations / (1e9 * 3600)  # 1B ops/sec

        print(f"{pruning_rate:8.0%} | {effective_combinations:14.2e} | {improvement_factor:12.1e} | {estimated_hours:13.2e} h")

    # 4. Complejidad espacial
    # bytes para solucion actual
    space_original = num_takes * 4
    # stack recursivo + estructuras
    space_improved = (num_takes ** 2) * 4 + num_takes * 40

    print(f"\nCOMPLEJIDAD ESPACIAL:")
    print(f"Algoritmo original: O(n) = {space_original} bytes")
    print(f"Algoritmo mejorado: O(n^2) = {space_improved} bytes")
    print(f"Overhead espacial: {space_improved/space_original:.1f}x")

    # 5. Complejidad por operacion
    print(f"\nCOMPLEJIDAD POR OPERACION:")

    operations_complexity = {
        'Preprocesamiento': ('O(n)', num_takes),
        'Ordenacion tomas': ('O(n log n)', num_takes * math.log2(num_takes)),
        'Poda por cota': ('O(1)', 1),
        'Estimacion optimista': ('O(n)', num_takes),
        'Evaluacion fitness': ('O(n)', num_takes),
        'Copia solucion': ('O(n)', num_takes)
    }

    for operation, (notation, value) in operations_complexity.items():
        print(f"{operation:20}: {notation:10} ≈ {value:8.1f} ops")

    # 6. Analisis asintotico para diferentes tamaños
    print(f"\nANALISIS ASINTOTICO:")
    print("Tomas | Comb. Originales | Con Poda 90% | Mejora Factor")
    print("-" * 60)

    test_sizes = [5, 10, 15, 20, 25, 30]
    avg_actors = np.mean(actors_per_take)

    for size in test_sizes:
        if size <= num_takes:
            original_complex = avg_actors ** size
            pruned_complex = original_complex * 0.1
            improvement = original_complex / pruned_complex
            print(f"{size:5d} | {original_complex:15.2e} | {pruned_complex:11.2e} | {improvement:12.1f}x")

    # 7. Metricas finales
    metrics = {
        'num_takes': num_takes,
        'total_combinations': total_combinations,
        'average_actors_per_take': np.mean(actors_per_take),
        'complexity_order': f"O({np.mean(actors_per_take):.1f}^n)",
        'space_complexity_original': space_original,
        'space_complexity_improved': space_improved,
        'theoretical_improvement_90pct': total_combinations / (total_combinations * 0.1)
    }

    return metrics


def benchmark_complexity_growth(base_matrix, max_takes=15):
    """
    Mide como crece la complejidad con el numero de tomas
    """
    print(f"\nBENCHMARK DE CRECIMIENTO DE COMPLEJIDAD")
    print("======================================")
    print("Tomas | Combinaciones | Tiempo Prep | Factor Crecimiento")
    print("-" * 60)

    previous_combinations = 1

    for n in range(3, min(max_takes + 1, len(base_matrix) + 1)):
        test_matrix = base_matrix[:n]

        start_time = time.time()
        combinations = 1
        for take in range(n):
            combinations *= sum(test_matrix[take])
        prep_time = time.time() - start_time

        growth_factor = combinations / previous_combinations if previous_combinations > 1 else 1

        print(f"{n:5d} | {combinations:12.2e} | {prep_time:10.6f}s | {growth_factor:17.2f}x")

        previous_combinations = combinations


def calculate_pruning_effectiveness(pruned_branches, total_evaluations):
    """
    Calcula la efectividad real de la poda basada en resultados de ejecucion
    """
    total_nodes = pruned_branches + total_evaluations
    pruning_rate = pruned_branches / total_nodes if total_nodes > 0 else 0

    print(f"\nEFECTIVIDAD REAL DE PODA:")
    print(f"Nodos evaluados: {total_evaluations:,}")
    print(f"Nodos podados: {pruned_branches:,}")
    print(f"Total nodos visitados: {total_nodes:,}")
    print(f"Tasa de poda: {pruning_rate:.2%}")
    print(f"Factor de mejora: {1/(1-pruning_rate) if pruning_rate < 1 else float('inf'):.1f}x")

    return {
        'pruning_rate': pruning_rate,
        'improvement_factor': 1/(1-pruning_rate) if pruning_rate < 1 else float('inf'),
        'nodes_saved': pruned_branches
    }


# Funcion principal para ejecutar todos los analisis
def run_complexity_analysis(availability_matrix):
    """
    Ejecuta analisis completo de complejidad
    """
    print("ANALISIS COMPLETO DE COMPLEJIDAD")
    print("=" * 50)

    # Metricas basicas
    metrics = calculate_complexity_metrics(availability_matrix)

    # Benchmark de crecimiento
    benchmark_complexity_growth(availability_matrix)

    # Resumen final
    print(f"\nRESUMEN DE COMPLEJIDAD:")
    print(f"Problema: {metrics['num_takes']} tomas")
    print(f"Complejidad teorica: {metrics['complexity_order']}")
    print(f"Combinaciones totales: {metrics['total_combinations']:.2e}")
    print(f"Mejora esperada con poda 90%: {metrics['theoretical_improvement_90pct']:.1f}x")

    return metrics

# Ejecutar analisis
if 'availability_matrix' in locals():
    complexity_metrics = run_complexity_analysis(availability_matrix)
else:
    print("Definir availability_matrix primero")
```
